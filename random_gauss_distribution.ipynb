```python
import random
import numpy as np
import matplotlib.pyplot as plt

num = 10000
mu = 0
sigma = 1
noise = []

for i in range(num):
    noise.append(random.gauss(mu,sigma))

plt.plot(noise)
plt.show()

```

```python
x = np.linspace(-5*sigma,5*sigma,101,endpoint=True)

cdf = []
count = 0
for i in x:
    for value in noise:
        if value<=i:
            count+=1
    cdf.append(count*101)
    count=0

plt.plot(cdf)
plt.show
        
    
```

```python
pdf=[]
count=0
for i in x:
    for value in noise:
        if i<=value<i+0.01:
            count+=1
    pdf.append(count/num)
    count=0

plt.plot(pdf)
plt.show
```
