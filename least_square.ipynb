```python
import numpy as np
import matplotlib.pyplot as plt

n = 51
x = np.linspace(0,1,n)

#linear
m = 4
c = 2
sigma = 2
y_noisy  = m*x + c + sigma * np.random.randn(1,n)
y_actual  = m*x + c 


plt.scatter(x,y_noisy,color='r')
plt.plot(x,y_actual)
```

```python
#least square
hT = np.vstack((np.ones(n),x))
print(hT)
A = np.matmul(np.linalg.inv(np.matmul(hT,np.transpose(hT))),hT)
print(A)
print(y_noisy.shape)
theta = np.matmul(A,np.transpose(y_noisy))
print(theta)

y_pred = np.matmul(np.transpose(theta) , hT)
print(y_pred)
```

```python
plt.scatter(x,y_noisy,color='r')
plt.plot(x,y_actual)
plt.plot(x,y_pred.reshape(-1,1),color='m')
```
