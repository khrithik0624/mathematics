```python
import numpy as np
import matplotlib.pyplot as plt

mu1 = 0
sigma1 = 1

x1= np.linspace(-5, 5, 1001, endpoint=True)
gaussian1 = 1/((2*np.pi)**0.5*sigma1)*np.exp(-0.5*(x1-mu1)**2/sigma1**2)

plt.plot(x1,gaussian1)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
 
def gaussian(x,mu,sigma):
    return 1/((2*np.pi)**0.5*sigma)*np.exp(-0.5*(x-mu)**2/sigma**2)
    
X1, X2 = np.meshgrid(x1, x1)

joint_gauss_2 = gaussian(X1,0,3)  * gaussian(X2,0,4)

fig = plt.figure()
ax= fig.add_subplot(111,projection='3d')
ax.plot_surface(X1,X2,joint_gauss_2,cmap='rainbow')
plt.show()

```
