```python
import numpy as np
import matplotlib.pyplot as plt


r = np.arange(0, 2, 0.01)
theta = 2 * np.pi * r

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(theta, r)
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```

```python
#circle
theta = np.linspace(0,2*np.pi,101,endpoint=True)
radius = 5
x = radius * np.cos(theta)
y = radius * np.sin(theta)
plt.axis('equal')
plt.plot(x,y) 
```

```python
#ellipse
a=4
b=3
x = a*np.cos(theta)
y = b*np.sin(theta)
plt.axis('equal')
plt.plot(x,y)
```

```python
#spiral
t = np.linspace(0,1,101,endpoint=True)
theta = t*6*np.pi
x = t*np.cos(theta)
y = t*np.sin(theta)
plt.axis('equal')
plt.plot(x,y)
```

```python
#spiral
t = np.linspace(0,6.5,1001,endpoint=True)
theta = t*2*np.pi
r = t**3 -5*t**2 + 3*t
x = r*np.cos(theta)
y = r*np.sin(theta)
plt.axis('equal')
plt.plot(x,y)
```

```python
#spiral
t = np.linspace(0,4,1001,endpoint=True)
theta = t*2*np.pi
r = np.exp(-t)
x = r*np.cos(theta)
y = r*np.sin(theta)
plt.axis('equal')
plt.plot(x,y)
```
