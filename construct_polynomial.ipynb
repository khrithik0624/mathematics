```python
import numpy as np
import matplotlib.pyplot as plt

roots = [0,1]
coeffs = []

def coeff(sum,i,n):
    if n<1:
        return sum
    if i>len(roots)-1:
        return 0
    sum = coeff(roots[i]*sum,i+1,n-1) + coeff(sum,i+1,n)
    return sum

minus_one = 1
for n in range(len(roots)+1):
    coeffs.append(coeff(1,0,n) * minus_one)
    minus_one = minus_one*-1

print(coeffs)

```

```python
x = np.linspace(0,3,1001,endpoint=True)

function_root = 1
for i in range(len(roots)):
    function_root = function_root*(x - roots[i])

plt.plot(x,function_root)

    
```

```python
polynomial = 0 

for i in range(len(coeffs)):
    polynomial = polynomial + x**(len(roots)-i) *coeffs[i]

plt.plot(x,polynomial)
```
