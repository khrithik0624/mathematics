```python
import matplotlib.pyplot as plt
import random
from matplotlib.animation import FuncAnimation

# Define the vertices of the equilateral triangle
vertices = [(0, 0), (10, 20), (20, 0)]

# Initialize the starting point
x, y = random.choice(vertices)

# Number of iterations
iterations = 100000

# Lists to store the x and y coordinates of points
x_points = [x]
y_points = [y]

# Perform the chaos game iterations
for _ in range(iterations):
    vertex = random.choice(vertices)
    x = (x + vertex[0]) / 2
    y = (y + vertex[1]) / 2
    x_points.append(x)
    y_points.append(y)

# Plot the points
plt.scatter(x_points, y_points, s=1, c='b', marker='.')
plt.title('Chaos Game')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
